```python
#df2=df.groupby(['CurrentState','yearOfStart']).size()
#df3=pd.pivot_table(df,index=['CurrentState','yearOfStart'],aggfunc=np.size)
#df3 
plt.figure(figsize=(8,6))
df4=df[df.CurrentState=='In'].groupby('yearOfStart').size().plot(color='blue')
df5=df[df.CurrentState=='Out'].groupby('yearOfStart').size().plot(color='red')
plt.xlim(1997,2014)
plt.xlabel("years of start")
plt.ylabel("count per category (current state In-blue- or out-red-)")
plt.show()
```
