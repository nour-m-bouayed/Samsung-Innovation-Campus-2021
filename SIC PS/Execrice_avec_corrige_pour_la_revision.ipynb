## Chi-Square Test

The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables.

```python
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as st
dataset=sns.load_dataset('tips')
```

```python
dataset.tail(5)
```

```python
dataset_table=pd.crosstab(dataset['sex'],dataset['smoker'])  
#crosstab : Compute a simple cross tabulation of two (or more) factors. By default
#computes a frequency table of the factors unless an array of values and an
#aggregation function are passed.
print(dataset_table)
```

```python
#we will try to find if there is any association between smoker and sex!!
# dataset_table.values print the observed values 
Observed_Values = dataset_table.values 
print("Observed Values :-\n",Observed_Values)
dataset_table.values
```

Question 2 

Use stats.chi2_contingency function to compute the chi-square statistic for the hypothesis test. Then compare it with the observed values!! Can you draw a conclusion?? 

```python
val=stats.chi2_contingency(dataset_table)
val
```

```python
val[3]
```

Now, let's define out ddof and alpha.

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:",ddof)
alpha = 0.05
```


Question 3

Use the following formula to calculate Z-score!!

$$
\chi^{2}=\sum \frac{(o-e)^{2}}{\ell}
$$

```python
from scipy.stats import chi2
#please use zip to loop through Observed_Values and Expected_Values
Expected_Values=val[3]
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
print("chi-square statistic:-",chi_square_statistic)
```

Question 4

What would be the result of percent point function ?

```python
#in the ppf function p value is assigned as alpha and q value as 1-alpha
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

From the critical_value and chi-square statistic we can define if there is a relationship between two categorical feature.


```python

#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

An F-test is used to test whether two population variances are equal. The null and alternative hypotheses for the test are as follows:

H0: σ1 = σ2 (the population variances are equal)

H1: σ1 ≠ σ2 (the population variances are not equal)

# Question 5

Knowing that: f-test is equel to $ \mathrm{s}_{1}^{2} / \mathrm{s}_{2}^{2}$.
Define a function to perform F-test in order to determine if the two populations samples x and y defined bellow have equal variances

```python
x = [18, 20 ,22 ,26, 28, 33, 41, 45, 51, 55]
y = [14, 15, 15, 17, 18, 22, 25, 25, 27, 34]
import numpy as np

#define F-test function
def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
    dfn = x.size-1 #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-scipy.stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
    return f, p

#perform F-test
f_test(x, y)

(4.38712, 0.019127)
```
