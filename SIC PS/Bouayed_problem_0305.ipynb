### "Correlation and interval estimation"

```python
import numpy as np
import scipy.stats as st
```

#### Answer the following questions by providing Python code:

Calculate the 95% and 99% conficence intervals for the Pearson correlation. <br>
Discuss whether the confidence interval includes zero or not.

1). When the sample size $n = 30$ and the sample correlation $r = 0.45$:

```python
r=0.45
n=30
z=np.arctanh(r)
se=1/np.sqrt(n-3)
```

```python
# 95% confidence interval. 
ci={'low':np.tanh(z-(-st.norm.ppf((1-0.95)/2))*se),'up':np.tanh(z+(-st.norm.ppf(1-0.95)/2)*se)}
print("95% Confidence interval",ci)
```

>The interval does nort include 0

```python
# 99% confidence interval. 
ci={'low':np.tanh(z-(-st.norm.ppf((1-0.99)/2))*se),'up':np.tanh(z+(-st.norm.ppf(1-0.99)/2)*se)}
print("99% Confidence interval",ci)
```

The interval includes zero 

2). When the sample size $n = 300$ and the sample correlation $r = -0.2$:

```python
r=-0.2
n=300
z=np.arctanh(r)
se=1/np.sqrt(n-3)
```

```python
# 95% confidence interval. 
ci={'low':np.tanh(z-(-st.norm.ppf((1-0.95)/2))*se),'up':np.tanh(z+(-st.norm.ppf(1-0.95)/2)*se)}
print("95% Confidence interval",ci)
```

>The interval does not include zero

```python
# 99% confidence interval. 
ci={'low':np.tanh(z-(-st.norm.ppf((1-0.99)/2))*se),'up':np.tanh(z+(-st.norm.ppf(1-0.99)/2)*se)}
print("99% Confidence interval",ci)
```

The interval does not include zero

####Conclusion : 
Even if we get a good sample correlation coefficient (relatively high), when the sample size is not big enough the confidence interval will be wide and the population r will be possibly low.

But if we have a large sample size, our confidence interval won't be as wide as in the first case
