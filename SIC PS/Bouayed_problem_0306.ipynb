### "F test of variance ratio"

```python
import numpy as np
import scipy.stats as st
```

#### Answer the following questions by providing Python code:

In a factory there are two packaging machines. Output samples are drawn from each machine. <br>
- $n_1=15$, $\bar{x}_1=5.0592 kg$, $s_1^2=0.1130 kg^2$
- $n_2=12$, $\bar{x}_2=4.9808 kg$, $s_2^2=0.0152 kg^2$  

Test whether there is a significant difference in the variances (95%).

```python
n1 = 15
ssq1 = 0.1130
n2 = 12
ssq2 = 0.0152
```

1). Do the right tail test:

```python
Ftest=ssq1/ssq2
1-st.f.cdf(Ftest,n1-1,n2-1)
```

2). Do the two tail test:

```python
Ftest=ssq1/ssq2
(1-st.f.cdf(Ftest,n1-1,n2-1))*2
```

3). Briefly discuss the results:

There is a very low probability (0.001969539320680891<0.2%) to observe these samples variances if populations variances are equal
Thus, we can reject the null hypothesis of equal variances in favour of the alternative hypothisis of variance 1 being greater than variance 2
