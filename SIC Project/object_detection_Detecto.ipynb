```python
from detecto import core, utils, visualize
from detecto.visualize import show_labeled_image, plot_prediction_grid
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
```

augmentation

```python
custom_transforms = transforms.Compose([
transforms.ToPILImage(),
transforms.Resize(900),
transforms.RandomHorizontalFlip(0.5),
transforms.ColorJitter(saturation=0.2),
transforms.ToTensor(),
utils.normalize_transform(),
])
```

training 

```python
Train_dataset=core.Dataset('train',transform=custom_transforms)#L1
Test_dataset = core.Dataset('test')#L2
loader=core.DataLoader(Train_dataset, batch_size=2, shuffle=True)#L3
model = core.Model(['Blueberry leaf', 'Tomato Septoria leaf spot', 'Corn leaf blight', 'Corn rust leaf', 'Apple Scab Leaf', 'Tomato leaf late blight', 'Tomato leaf yellow virus', 'Bell_pepper leaf spot', 'Tomato mold leaf', 'Potato leaf early blight', 'Bell_pepper leaf', 'Tomato leaf bacterial spot', 'Potato leaf late blight', 'Raspberry leaf', 'grape leaf black rot', 'Squash Powdery mildew leaf', 'Apple rust leaf', 'Tomato leaf mosaic virus', 'Tomato Early blight leaf', 'Tomato leaf', 'Soyabean leaf', 'Peach leaf', 'Soybean leaf', 'Corn Gray leaf spot', 'Apple leaf', 'Strawberry leaf', 'Cherry leaf', 'grape leaf', 'Potato leaf', 'Tomato two spotted spider mites leaf'])#L4
losses = model.fit(loader, Test_dataset, epochs=25, lr_step_size=5, learning_rate=0.001, verbose=True)#L5
```

```python
plt.plot(losses)
plt.show()
```

Model saving, loading, and predicting

```python
model.save(‘model_weights.pth’)
model = core.Model.load(‘model_weights.pth’, [‘Wheel’, ‘Head Light’])
```

```python
image = utils.read_image(‘Test/Maruti car_27.jpeg’) 
predictions = model.predict(image)
labels, boxes, scores = predictions
show_labeled_image(image, boxes, labels)
```
