```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from os import listdir
train=listdir('PlantDoc-Dataset/train')
values=[]
for cat in train:
    for img in listdir(f'PlantDoc-Dataset/train/{cat}'):
        value=(img,cat)
        values.append(value)
train_df=pd.DataFrame(values,columns=['filename','class_name'])
```

```python
from os import listdir
test=listdir('PlantDoc-Dataset/test')
values=[]
for cat in test:
    for img in listdir(f'PlantDoc-Dataset/test/{cat}'):
        value=(img,cat)
        values.append(value)
test_df=pd.DataFrame(values,columns=['filename','class_name'])
```

```python
species=[s.split()[0] for s in train_df['class_name'].values]
train_df['species']=species
```

```python
train_df.head()
```

```python
species=[s.split()[0] for s in test_df['class_name'].values]
test_df['species']=species
```

```python
sns.countplot(x=species,data=train_df)
plt.xticks(rotation=50)
plt.title('distribution of species in training set_plantdoc')
```

```python
sns.countplot(x=species,data=test_df)
plt.xticks(rotation=50)
plt.title('distribution of species in testing set_ plantdoc')
```

```python
sns.countplot(x='class_name',data=train_df[train_df.species=='Tomato'])
plt.xticks(rotation=90)
plt.title('distribution of classes in tomatoes')
```

```python
sns.countplot(x='class_name',data=test_df[test_df.species=='Tomato'])
plt.xticks(rotation=90)
plt.title('distribution of classes in tomatoes (Plantdoc testing set)')
```
