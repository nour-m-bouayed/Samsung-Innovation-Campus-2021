#TRAINING SET

```python
train_path = '/content/PlantDiseasesDataset/Tomato/train'
valid_path = '/content/PlantDiseasesDataset/Tomato/valid'
```

```python
import os 
import pandas as pd
train=os.listdir(train_path)
train_df=pd.DataFrame(columns=['class_name','number of images'])
for c in train:
    train_df.loc[train_df.shape[0]]=[c,len(os.listdir(f'{train_path}/{c}'))]
```

```python
train_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='class_name',y='number of images',data=train_df)
plt.title('distribution of Tomato images over classes of disease in PlantifyDr training set')
plt.xticks(rotation=90)
```

```python
import os 
import pandas as pd
valid=os.listdir(valid_path)
valid_df=pd.DataFrame(columns=['class_name','number of images'])
for c in valid:
    valid_df.loc[valid_df.shape[0]]=[c,len(os.listdir(f'{valid_path}/{c}'))]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='class_name',y='number of images',data=valid_df)
plt.title('distribution of Tomato images over classes of disease in PlantifyDr validation set')
plt.xticks(rotation=90)
```
