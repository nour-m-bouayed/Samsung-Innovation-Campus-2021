#### Assignment: 
- Write a function named 'palindrome' to test whether the entered word or sentence is a palindrome.<br>
- Ignore capitalization, spaces, punctuation marks, symbols, etc. <br>
- Use the properties of 'queue' and 'stack' data structures. <br>

Palindrome examples: <br>
- mom
- wow
- noon
- level
- radar
- kayak
- racecar
- God's dog
- Madam, I'm Adam

```python
from re import sub

#using a double_ended queue represented by a list : 

def palindrome(d):
  d=sub("[,.: ']","",d)
  d=d.lower()
  letters=[c for c in d]
  result=True
  while (len(letters)>1):
    #we dequeue the list from the left and from the right simultaneously and we compare the two characters
    if letters.pop(-1)!=letters.pop(0):
      result=False
  return result   
    
    
    
```

Test results:

```python
palindrome("Kayak")
```
