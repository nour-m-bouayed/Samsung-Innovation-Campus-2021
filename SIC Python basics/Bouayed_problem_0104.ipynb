## Quiz #0104 

#### Code the class 'Aquarium' that outputs the result as shown. 

```python
class Aquarium:
  def __init__(self):
    print("An Aquarium has opened!")
    self.list_animals=[] 
  def enters(self,animal):
    if animal not in self.list_animals:
      print(f"{animal} is a new member of our Aquarium")
      self.list_animals.append(animal)
    else:
      print(f"We already have {animal}")
  def show_all(self):
    print(self.list_animals)
  def leaves(self,animal):
    if animal not in self.list_animals:
      print("No such sea creature in the Aquarium!")
    else:
      self.list_animals.remove(animal)  
    
```
