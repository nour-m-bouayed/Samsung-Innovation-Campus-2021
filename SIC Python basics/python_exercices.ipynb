# Exercice 1

Write a Python function to find the Max of three numbers.

````python
def maximum(a,b,c):
  m=a
  l=[b,c]
  for i in l:
    if m<i:
      m=i 
  return m

print(maximum(-1,2,0))
````

# Exercice 3

Write a Python function that takes a list and returns a new list with unique elements of the first list.



```
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```



````python
def distinct(lst):
  l=[]
  for i in lst:
    if i not in l:
      l.append(i)
  return l
sample_list=[1,2,3,3,3,3,4,5]
unique_list=distinct(sample_list)
print(unique_list)
````

# Exercice 7

Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

````python
def upper_lower(s):
  upper=0
  lower=0
  for i in s:
    if (ord(i)<=122 and ord(i)>=97):
      lower+=1
    if (ord(i)>=65 and ord(i)<=90):
      upper+=1
  return (upper,lower)

print(upper_lower("NoUr MeRiEm"))
````

# Exercice 17
Write the definition of a Point class. Objects from this class should have a

- a method show to display the coordinates of the point
- a method move to change these coordinates.
- a method dist that computes the distance between 2 points.
Note the distance between 2 points A(x0, y0) and B(x1, y1) can be compute as follows
$d(A,B)=\sqrt{(x_{1}-x_{0})^{2} + (y_{1}-y_{0})^{2} }$

this is the expected result

```
p1 = Point(2, 3)
>>> p2 = Point(3, 3)
>>> p1.show()
(2, 3)
>>> p2.show()
(3, 3)
>>> p1.move(10, -10)
>>> p1.show()
(12, -7)
>>> p2.show()
(3, 3)
>>> p1.dist(p2)
1.0
```



````python
import math as m
class Point:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def show(self):
    print((self.x,self.y))
  def move(self,xT,yT):
    self.x+=xT
    self.y+=yT
  def dist(self,p):
    d=m.sqrt((p.x-self.x)**2+(p.y-self.y)**2)
    return d

p1=Point(2,3)
p2=Point(3,3)
p1.show()
p2.show()
p1.move(10,-10)
p1.show()
p1.dist(p2)
````

````python
class Matrix:
  def __init__(self,row,col,val=None):
    self.row=row
    self.col=col
    self.val=val
    self.l=[[val]*col for x in range(row)]
  def __str__(self):
    s=""
    for i in range(self.row):
      for j in range(self.col):
        s+=str(self.l[i][j])+"  "
      s+="\n"
    return (s)
  def size(self):
    return (self.row,self.col)
  def get_cell(self,row,col):
    return self.l[row][col]
  def set_cell(self,row,col,val):
    self.l[row][col]=val
  def to_str(self):
    s=""
    for i in range(self.row):
      for j in range(self.col):
        s+=str(self.l[i][j])+"  "
      s+="\n"
    return (s)
  def mult(self,k):
    m=Matrix(self.row,self.col)
    m.l=[self.l[x].copy() for x in range(self.row)]
    for i in range(m.row):
      for j in range(m.col):
        m.l[i][j]*=k
    return (m)
m1=Matrix(2,2,2)
print(m1)
print(m1.mult(2))
print(m1)
````

````python
#Write a python program to find the longest words
def max_len(lst):
  l=list(map(lambda x:len(x),lst))
  return (list(filter(lambda x:len(x)==max(l),lst)))

with open("/basic_file.txt","r")as f:
  words=f.read().split()
  print(words)
  print(max_len(words))
````

````python
#Write a Python program to count the frequency of words in a file.
def frequencies(file):
  with open(file,"r")as f:
    frequencies={}
    words=f.read().split()
    for word in words:
      if word not in frequencies.keys():
        frequencies[word]=1
      else:
        frequencies[word]+=1
  return (frequencies)
print(frequencies("/basic_file.txt"))
````
