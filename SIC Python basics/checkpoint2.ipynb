```python
def squares(n):
  dict={}
  for i in range (1,n+1):
    dict[i]=i*i
  return(dict)
n=int(input("please enter a number"))
dico=squares(n)
print(dico)
```

```python
def missing_char(str,n):
  str2=str[:n]+str[n+1:]
  return str2

str=input("enter a string")
n=int(input("enter the position of the letter u want to remove"))
result=missing_char(str,n)
print(result)
```
