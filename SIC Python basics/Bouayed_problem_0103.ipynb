## Quiz #0103 

#### Code the class named 'Calculator' that outputs the result as shown. 

```python
class Calculator:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def sum(self):
    return self.a+self.b
  def subtract(self):
    return self.a-self.b
  def product(self):
    return self.a*self.b
  def divide(self):
    return self.a/self.b
```

Test results:

```python
calc1 = Calculator(5,3)
calc2 = Calculator(10,2)
```

```python
print(calc1.sum())
print(calc1.subtract())
print(calc1.product())
print(calc1.divide())
```

```python
print(calc2.sum())
print(calc2.subtract())
print(calc2.product())
print(calc2.divide())
```
