```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

#### Read in data:

```python
# Load data.
data = load_breast_cancer()
# Explanatory variables.
X = data['data']
# Relabel such that 0 = 'benign' and 1 = malignant.
Y = 1 - data['target']
```

```python
# Split the dataset into training and testing.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=1234)
```

1). Define the 'sigmoid' and 'gradient' functions to produce the output shown below:

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def gradient(X, Y, beta):
  n=len(X)
  return (np.dot(X.T,(sigmoid(np.dot(X,beta))-Y)))/n
       # <Your code goes in here>
```

2). Define the 'LogisticRegression' class to produce the output shown below:

```python
class LogisticRegression:
    def __init__(self, learn_rate):
        self.rate = learn_rate
        self.n_nodes = None
        self.beta = None
        
    def train(self, input_X, input_Y, n_epochs):
      X,n_samples,n_features=self._add_ones(input_X)
      #init the weights with zeros
      self.beta=np.zeros(n_features)
      #gradient descent
      for _ in range(n_epochs):
        self.beta-=self.rate*gradient(X,input_Y,self.beta)
    
    def query(self, input_X, cutoff=0.5):
       X,n_samples,n_features=self._add_ones(input_X)
       #Y_predicted
       Y_predicted=sigmoid(np.dot(X,self.beta))
       Y_predicted_classes=[1 if i>cutoff else 0 for i in Y_predicted]
       return np.array(Y_predicted_classes)
       
    #utility funciton
    def _add_ones(self,input_X):
      n_samples,n_features=input_X.shape
      ones=np.ones((n_samples,1))
      X=np.concatenate((input_X,ones),axis=1)
      n_features+=1
      return X,n_samples,n_features
             
```

#### Sample run:

```python
# Hyperparameter for the learner.
learning_rate = 0.001
```

```python
# Train and predict.
LR = LogisticRegression(learning_rate)
LR.train(X_train, Y_train, 2000)
Y_pred = LR.query(X_test,cutoff=0.5)
```
