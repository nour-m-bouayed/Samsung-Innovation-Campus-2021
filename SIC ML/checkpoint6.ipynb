### b) splitting our data into training dataset

```python
import random

def loadDataset(filename, split, trainingSet=[] , testSet=[]):
  with open(filename, 'r') as csvfile:
    lines = csv.reader(csvfile)
    dataset = list(lines)
    for x in range(len(dataset)):
      for y in range(4):
        dataset[x][y] = float(dataset[x][y])
      if random.random() < split:
        trainingSet.append(dataset[x])
      else:
        testSet.append(dataset[x])
```

```python
trainingSet=[]

testSet=[]

loadDataset('iris.data.txt', 0.66, trainingSet, testSet)

print ('Train: ' + repr(len(trainingSet)))

print ('Test: ' + repr(len(testSet)) )
```

##2. Similarity

Given that all four flower measurements are numeric and have the same units, we can directly use the Euclidean distance measure. 



```python
import math

def euclideanDistance(instance1, instance2, length):
  return (math.sqrt(sum([math.pow((instance1[i]-instance2[i]),2) for i in range(length)])))
```

```python
data1 = [2, 2, 2, 'a']

data2 = [4, 4, 4, 'b']

distance = euclideanDistance(data1, data2, 3)

print('Distance: ' + repr(distance))
```

##3. Neighbors

```python
import operator

def getNeighbors(trainingSet, testInstance, k):
  distances = []
  length = len(testInstance)-1
  for x in range(len(trainingSet)):
    dist = euclideanDistance(testInstance, trainingSet[x], length)
    distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1))#sort by ascending order of the 2nd item 'dist' of each tuple of distances 
  neighbors = []
  for x in range(k):
    neighbors.append(distances[x][0])
  return neighbors
```

```python
trainSet = [[2, 2, 2, 'a'], [4, 4, 4, 'b']]

testInstance = [5, 5, 5]

k = 1

neighbors = getNeighbors(trainSet, testInstance, 1)

print(neighbors)
```

##4. Response

```python
import operator

def getResponse(neighbors):
  classVotes = {}
  for x in range(len(neighbors)):
    response = neighbors[x][-1] #complete with appropriate number
    if response in classVotes:
      classVotes[response]+=1
    else:
      classVotes[response]=1
  sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
  return sortedVotes[0][0]
```

```python
#test the function
neighbors = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]

response = getResponse(neighbors)

print(response)
```

##5. Accuracy

```python
def getAccuracy(testSet, predictions):
  correct=0
  for actual,predicted in zip(testSet,predictions):
    if actual[-1]==predicted:
      correct+=1
  return (correct/float(len(testSet))) * 100.0
```

```python
testSet = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]

predictions = ['a', 'a', 'a']

accuracy = getAccuracy(testSet, predictions)

print(accuracy)
```

##6. Main

```python
def main():
  trainingSet = []
  testSet = []
  split = 0.7
  loadDataset('iris.data.txt', split, trainingSet, testSet)
  predict = []
  k = 4
  for x in range(len(testSet)):
    neighbors = getNeighbors(trainingSet, testSet[x], k)
    response = getResponse(neighbors)
    predict.append(response)
    
  accuracy = getAccuracy(testSet, predict)
  print('Accuracy: ',accuracy,'%')
```

```python
main()
```

##7. Another distance metric

Manhattan distance : 

```python
def manhattanDistance(instance1, instance2, length):
  return (sum([abs(instance1[i]-instance2[i]) for i in range(length)]))
```

```python
#testing it 
data1 = [2, 2, 2, 'a']

data2 = [4, 4, 4, 'b']

distance = manhattanDistance(data1, data2, 3)

print('Distance: ' + repr(distance))
```
